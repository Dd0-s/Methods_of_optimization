## Задача 1

Рассмотрим билинейную седловую задачу:
\begin{equation}
\min_{x \in [-10; 10]^d} \max_{y \in [-10; 10]^d} f(x,y) = (x-b_x)^T A (y - b_y) + \frac{\lambda}{2}\| x - b_x\|^2 - \frac{\lambda}{2}\| y - b_y\|^2,
\end{equation}
где в качестве матрицы $A$ - случайно сгенерированная матрица с $L =1000$, $\mu = 0.01$, $d = 100$, $b_x$ и $b_y$ - случайные векторы c компонентами из отрезка $[-10; 10]$, $\lambda = 1$.

__(а)__ Какое решение у данной задачи? Является ли данная задача сильно выпуклой по $x$ и сильно вогнутой по $y$? Если да, то чему равна константа сильной выпуклости? Является ли оператор, соотвествующего это задаче вариационного неравенства, сильно монотонным? Липшицевым? Чему равны $L$ и $\mu$ для данного оператора? 

__1. Является ли данная задача сильно выпуклой по $x$ и сильно вогнутой по $y$?__

$$ (x - b_x)^T A (y - b_y) = \langle x - b_x, A (y - b_y) \rangle = \langle A^T (x - b_x), y - b_y \rangle$$

$$ \| x - b_x \|^2 = \langle x - b_x, x - b_x \rangle$$

$$ \| y - b_y \|^2 = \langle y - b_y, y - b_y \rangle$$

Тогда:

$$ \nabla_x f(x, y) = A (y - b_y) + \lambda (x - b_x) $$

$$ \nabla_x^2 f(x, y) = \lambda I $$

$$ \nabla_y f(x, y) = A^T (x - b_x) - \lambda (y - b_y) $$

$$ \nabla_y^2 f(x, y) = -\lambda I $$

Получаем, что задача сильно выпукла по $x$ при $\mu_x = \lambda$ и сильно вогнута по $y$ при $\mu_y = \lambda$. То есть $\mu_x = \mu_y = \lambda$.

__2. Если да, то чему равна константа сильной выпуклости?__

Константа сильной выпуклости / вогнутости равна $\lambda$.

__3. Какое решение у данной задачи?__

Так как задача сильно выпукла по $x$ и сильно вогнута по $y$, то:

$$ \nabla_x f(x, y) = A (y^* - b_y) + \lambda (x^* - b_x) = 0 $$

$$ \nabla_y f(x, y) = A^T (x^* - b_x) - \lambda (y^* - b_y) = 0 $$

Причем существует единственное решение:

$$ y^* = b_y $$

$$ x^* = b_x $$

__4.Является ли оператор, соотвествующего это задаче вариационного неравенства, сильно монотонным?__

Возьмём: 

$$z = \left( 
          \begin{array}{cc}
              x \\
              y 
          \end{array}
      \right)
$$

$$F(z) = \left( 
                \begin{array}{cc}
                    \nabla_x f(x, y) \\
                    -\nabla_y f(x, y) 
                \end{array}
            \right)
$$

Определение $\mu$ - сильной монотонности:

$$ \langle F(z_1) - F(z_2), z_1 - z_2 \rangle \leq \mu \|z_1 - z_2\|_2^2 $$

$$ \langle F(z_1) - F(z_2), z_1 - z_2 \rangle = \langle \nabla_x f(x_1, y_1) - \nabla_x f(x_2, y_2), x_1 - x_2 \rangle - \langle \nabla_y f(x_1, y_1) - \nabla_y f(x_2, y_2), y_1 - y_2 \rangle$$

$$ \langle \nabla_x f(x_1, y_1) - \nabla_x f(x_2, y_2), x_1 - x_2 \rangle = \langle A (y_1 - y_2), x_1 - x_2 \rangle + \lambda \| x_1 - x_2 \|_2^2$$

$$ \langle \nabla_y f(x_1, y_1) - \nabla_y f(x_2, y_2), y_1 - y_2 \rangle = \langle A^T (x_1 - x_2), y_1 - y_2 \rangle - \lambda \| y_1 - y_2 \|_2^2$$

Тогда:

$$ \langle F(z_1) - F(z_2), z_1 - z_2 \rangle = \lambda \left( \| x_1 - x_2 \|_2^2 + \| y_1 - y_2 \|_2^2 \right) = \lambda \|z_1 - z_2\|_2^2$$

То есть $\mu = \lambda$.

__4. Липшицевым?__

Определение липшивости:

$$ \| F(z_1) - F(z_2) \|_2 \leq L \|z_1 - z_2\|_2 $$

$$ \| F(z_1) - F(z_2) \|_2 \leq \| \nabla_x f(x_1, y_1) - \nabla_x f(x_2, y_2) \|_2 + \| \nabla_y f(x_1, y_1) - \nabla_y f(x_2, y_2) \|_2 $$

$$ \| \nabla_x f(x_1, y_1) - \nabla_x f(x_2, y_2) \|_2 = \| A (y_1 - y_2) + \lambda (x_1 - x_2) \|_2 \leq \| A \|_2 \| y_1 - y_2 \| + \lambda \| x_1 - x_2 \|$$

$$ \| \nabla_y f(x_1, y_1) - \nabla_y f(x_2, y_2) \|_2 = \| A^T (x_1 - x_2) - \lambda (y_1 - y_2) \|_2 \leq \| A^T \|_2 \| x_1 - x_2 \| + \lambda \| y_1 - y_2 \|$$

Так как $\| A \|_2 = \| A^T \|$, то можем записать:

$$ \| F(z_1) - F(z_2) \|_2 \leq (\| A \|_2 + \lambda)(\| x_1 - x_2 \|_2 + \| y_1 - y_2 \|_2) \leq 2 ( \| A \|_2 + \lambda ) \|z_1 - z_2\|_2$$

То есть:

$$L = 2 (\| A \|_2 + \lambda) = 2 \left( \sqrt{\lambda_{max}(A^T A)} + \lambda \right)$$

__5. Чему равны $L$ и $\mu$ для данного оператора?__

$L$ мы нашли в предыдущем пункте __4.__: 

$$L = 2 \left( \sqrt{\lambda_{max}(A^T A)} + \lambda \right) $$

$L$ мы нашли в предыдущем пункте __3.__:

$$\mu = \lambda$$

__(б)__ Решите данную задачу с помощью метода градиентного спуска(подъема) с проекцией и с помощью метода Экстраградиента с проекцией(смотри Секцию 3.3 [статью](https://arxiv.org/pdf/1802.10551.pdf)). Для начала попробуйте теоретические шаги для данных методов. Далее подберите шаг для наилучшей сходимости. Постройте график сходимости (относительное расстояние до решения по аргументу от числа подсчетов градиентов). На графике должно быть 4 линии, по две на каждый из методов. Сделайте вывод.

Формула для поиска проекции:

$$\text{proj}(x) = \underset{t \in [-10; 10]}{\arg \min} \left( \frac{1}{2} \| t - x \|_2^2 \right)$$

1. Если $x < -10$, то $proj(x) = -10$
2. Если $-10 \geq x \leq 10$, то $proj(x) = x$
3. Если $x > 10$, то $proj(x) = 10$

```python
# Генерация матрицы
def gen_matrix(d, mu, L, s):
    np.random.seed(s)
    B = np.random.sample((d, d))
    Q, _ = np.linalg.qr(B)
    C = np.diag(mu + (L - mu) * np.random.sample(d))
    C[0][0] = L
    C[d - 1][d - 1] = mu
    A = Q @ C @ Q.T
    return A
```

```python
# Целевая функция
def f(x, y):
    return (x - b_x).T @ A @ (y - b_y) + 0.5 * lambd * norm(x - b_x, ord=2)**2 - 0.5 * lambd * norm(y - b_y, ord=2)**2

# Градиент функции по x
def grad_f_x(x, y):
    return A @ (y - b_y) + lambd * (x - b_x)

# Компонента градиента функции
def grad_f_y(x, y):
    return A.T @ (x - b_x) - lambd * (y - b_y)

# Проекция
def proj(x):
    s = np.zeros(len(x))
    for i, x_i in enumerate(x):
        if x_i < -10:
            s[i] = -10
        elif x_i > 10:
            s[i] = 10
        else:
            s[i] = x_i
    return s
```

```python
s = 2
L = 1000
mu = 0.01
d = 100
lambd = 1

np.random.seed(s)
A = gen_matrix(d, mu, L, s)
b_x = 20 * (np.random.sample(d) - 0.5)
b_y = 20 * (np.random.sample(d) - 0.5)

x_0 = np.random.sample(d)
y_0 = np.random.sample(d)

R = np.linalg.norm([x_0 - b_x, y_0 - b_y], ord=2)**2
```

Метод градиентного спуска(подъема):

$$ x^{k + 1} = \text{proj} \left( x^k - \gamma_k \nabla_x f(x^k, y^k) \right) $$

$$ y^{k + 1} = \text{proj} \left( y^k + \gamma_k \nabla_y f(x^k, y^k) \right) $$

Теоритический шаг GD равен:

$$ \gamma_k = \dfrac{\mu}{L^2} $$

Для подбора наилучшего шага рассмотрим:

$$ \gamma_k = \dfrac{c \mu}{L^2}, $$

затем выберем лучший.

```python
def DG(grad_f_x, grad_f_y, proj, x_0, y_0, gamma, iteration, epsilon=10**-7, c=1):
    
    x_now = x_0
    y_now = y_0
    errors = []
    number = [0]
    
    errors.append(np.linalg.norm([x_now - b_x, y_now - b_y], ord=2)**2 / R)
    
    for k in trange(iteration, colour='green'):
        
        x_next = proj(x_now - gamma(k, c) * grad_f_x(x_now, y_now))
        y_next = proj(y_now + gamma(k, c) * grad_f_y(x_now, y_now))
        
        x_now = x_next
        y_now = y_next
        
        errors.append(np.linalg.norm([x_now - b_x, y_now - b_y], ord=2)**2 / R)
        number.append(number[-1] + 2)

        if errors[-1] < epsilon:
            break
            
    return x_now, y_now, errors, number
```

```python
def gamma_GD(k, c=1):
    return c * mu / (L * L)
```

```python
_, _, errors_1, number_1 = DG(grad_f_x, grad_f_y, proj, x_0, y_0, gamma_GD, 10**5, c=1)
_, _, errors_2, number_2 = DG(grad_f_x, grad_f_y, proj, x_0, y_0, gamma_GD, 10**5, c=200)
_, _, errors_3, number_3 = DG(grad_f_x, grad_f_y, proj, x_0, y_0, gamma_GD, 10**5, c=400)
_, _, errors_4, number_4 = DG(grad_f_x, grad_f_y, proj, x_0, y_0, gamma_GD, 10**5, c=600)
_, _, errors_5, number_5 = DG(grad_f_x, grad_f_y, proj, x_0, y_0, gamma_GD, 10**5, c=800)
```

```python
plt.figure(figsize = (20, 10))
plt.plot(number_1, np.log(errors_1), "red", label = "c = 1")
plt.plot(number_2, np.log(errors_2), "cyan", label = "c = 200")
plt.plot(number_3, np.log(errors_4), "orange", label = "c = 400")
plt.plot(number_4, np.log(errors_3), "green", label = "c = 600")
plt.plot(number_5, np.log(errors_5), "purple", label = "c = 800")
plt.title("Логарифм значения критерия от числа подсчетов градиента", fontsize = 18)
plt.xlabel("Число подсчетов градиента", fontsize = 18)
plt.ylabel("Логарифм значения критерия", fontsize = 18)
plt.legend(fontsize = 20)
plt.show()
```

Метод экстраградиента:

$$ x^{k + 1 / 2} = \text{proj} \left( x^k - \gamma_k \nabla_x f(x^{k - 1 / 2}, y^{k - 1 / 2}) \right)$$

$$ x^{k + 1} = \text{proj} \left( x^k - \gamma_k \nabla_x f(x^{k + 1 / 2}, y^{k + 1 / 2}) \right)$$

$$ y^{k + 1 / 2} = \text{proj} \left( y^k + \gamma_k \nabla_y f(x^{k - 1 / 2}, y^{k - 1 / 2}) \right)$$

$$ y^{k + 1} = \text{proj} \left( y^k + \gamma_k \nabla_y f(x^{k + 1 / 2}, y^{k + 1 / 2}) \right)$$

Теоритический шаг EG равен:

$$ \gamma_k = \dfrac{1}{4L} $$

Для подбора наилучшего шага рассмотрим:

$$ \gamma_k = \dfrac{c}{4L} $$

затем выберем лучший.

```python
def EG(grad_f_x, grad_f_y, proj, x_0, y_0, gamma, iteration, epsilon=10**-7, c=1):
    
    x_now = x_0
    x_half_now = x_0
    y_now = y_0
    y_half_now = y_0
    errors = []
    number = [0]
    
    errors.append(np.linalg.norm([x_now - b_x, y_now - b_y], ord=2)**2 / R)
    
    for k in trange(iteration, colour='green'):
        
        x_half_next = proj(x_now - gamma(k, c) * grad_f_x(x_half_now, y_half_now))
        x_next = proj(x_now - gamma(k, c) * grad_f_x(x_half_now, y_half_now))
        y_half_next = proj(y_now + gamma(k, c) * grad_f_y(x_half_now, y_half_now))
        y_next = proj(y_now + gamma(k, c) * grad_f_y(x_half_now, y_half_now))
        
        x_now = x_next
        x_half_now = x_half_next
        y_now = y_next
        y_half_now = y_half_next
        
        errors.append(np.linalg.norm([x_now - b_x, y_now - b_y], ord=2)**2 / R)
        number.append(number[-1] + 4)

        if errors[-1] < epsilon:
            break
            
    return x_now, y_now, errors, number
```

```python
def gamma_EG(k, c=1):
    return c / (4 * L)
```

Теперь рассмотрим билинейную задачу на симплексе. Описание задачи возьмите из [лекций](https://www2.isye.gatech.edu/~nemirovs/BrazilTransparenciesJuly4.pdf) - Слайд 125 "How it Works: Policeman vs. Burglar" и следующий.

__(а)__ Сгенерируйте матрицу для данной задачи размера $n=10$ c $\theta = 1$.

Есть $n^2$ домов, с координатами $(i \text{ mod } n; i \text{ div } n)$, расположенных в квадрате $[0; n-1] \times [0; n-1]$.

Матрица $A$:

$$ A_{ij} = w_i (1 - \exp \left\{ -\theta \text{ dist}(i , j) \right\})$$

Парметры $w_i$ выбираются из соображений, что чем ближе дом к центру, тем коэффициент больше (константа нужна побольше, чтобы пирамида весов была более гладкой):

$$ w_i = \frac{1}{\text{dist}(i, i_{center}) + 30},$$

Minimax задача:

$$\min_{x \in \Delta_d} \max_{y \in \Delta_d} f(x,y) := y^T A x, \quad d = n^2. $$

Для нахождения $\mu$ и $L$ воспользуемся результатми, полученными в задаче 1, при условии: $b_x = b_y = 0$, $\lambda = 0$ и $A = A^T$: 

$$\mu = 0, \quad L = 2 \cdot \lambda_{max}(A)$$

```python
def f(x, y):
    return y.T @ A @ x

def grad_f_x(x, y):
    return A.T @ y

def grad_f_y(x, y):    
    return A @ x
```

```python
def dist(i, j):
    
    i_x = i % n
    i_y = i // n
    j_x = j % n
    j_y = j // n
    
    return np.linalg.norm([i_x - j_x, i_y - j_y], ord=np.inf)
```

```python
n = 10
d = n * n
theta = 1

i_center = (n // 2) * (n + 1)
w = np.array([1 / (dist(i, i_center) + 10) for i in range(d)])

A = np.zeros((d, d))
for i in range(d):
    for j in range(d):
        A[i][j] = w[i] * (1 - np.exp(-theta * dist(i, j)))

L = 2 * np.max(np.linalg.eig(A)[0]).real        
        
s = 50
np.random.seed(s)
x_0 = np.random.sample(d) / np.linalg.norm(x_0, ord=1)
y_0 = np.random.sample(d) / np.linalg.norm(y_0, ord=1)

print(f"L = {np.round(L, 2)}")
```

__(б)__ Для данной задачи предложите модификацию метода Экстраградиента, которая учитывает, что задача решается на симплексе. В этом может помочь [статья](https://arxiv.org/pdf/0809.0815.pdf) (смотри Секцию 3.3). Реализуйте данный метод.

Воспользуемся SMP:

$$ w_x^k = P_{r^{k - 1}_x}(\gamma_k \nabla_x f(r^{k - 1}_x, r^{k - 1}_y)) $$

$$ w_y^k = P_{r^{k - 1}_y}(-\gamma_k \nabla_y f(r^{k - 1}_x, r^{k - 1}_y)) $$

$$ r_x^k = P_{r^{k - 1}_x}(\gamma_k \nabla_x f(w^{k}_x, w^{k}_y)) $$

$$ r_y^k = P_{r^{k - 1}_y}(-\gamma_k \nabla_y f(w^{k}_x, w^{k}_y)), $$

где 

$$ P_z(t)_j := \frac{z_j \exp(-t_j)}{\sum\limits_{i = 1}^d z_i \exp(-t_i)} $$

Критерий сходимости:

$$ \text{Err}_{\text{vi}}(z) := \max_{u \in \Delta_d \times \Delta_d} \langle F(u), z - u \rangle $$

В нашем случае: 

$$z = \left(
          \begin{array}{cc}
              \hat{x} \\
              \hat{y}
          \end{array}
       \right),~
u = \left(
    \begin{array}{cc}
        x \\
        y \end{array}
    \right),~ 
F(u) = \left(
           \begin{array}{cc}
               A^T y \\
               -Ax
           \end{array}
       \right)
$$

Тогда:

$$ \text{Err}_{\text{vi}}(z) = \max_{x, y \in \Delta_d} \left\{ y^TA(\hat{x} - x) - x^TA^T(\hat{y} - y) \right\} = \max_{x, y \in \Delta_d} \left\{ y^TA\hat{x} - y^TAx - \hat{y}^TAx + y^TAx \right\} = \max_{x, y \in \Delta_d} \left\{ y^TA\hat{x} - \hat{y}^TAx \right\} $$

Функция Лагранжа:

$$ \mathcal{L}(x, y, \lambda_x, \lambda_y, \mu_x, \mu_y) = \hat{y}^TAx - y^TA\hat{x} - \lambda_x^Tx - \lambda_y^Ty + \mu_x \left( \sum_{i = 1}^d x_i - 1 \right) + \mu_y \left( \sum_{i = 1}^d y_i - 1 \right)$$

Условия ККТ:

$$ \nabla_x \mathcal{L} = A^T\hat{y} - \lambda_x + \mu_x \cdot \textbf{1} = 0 $$

$$ \nabla_y \mathcal{L} = -A\hat{x} - \lambda_y + \mu_y \cdot \textbf{1} = 0 $$

$$ \lambda_x^T x = 0, ~\lambda_x \succeq 0 $$

$$ \lambda_y^T y = 0, ~\lambda_y \succeq 0 $$

$$ \sum_{i = 1}^d y_i = 1 $$

$$ \sum_{i = 1}^d x_i = 1 $$

Обозначим: 

$$ j_x = \arg\min_{j \in \overline{1, d}} (A^T\hat{y})_j $$

$$ j_y = \arg\max_{j \in \overline{1, d}} (A\hat{x})_j $$

Тогда решение:

$$ \mu_x = -(A^T\hat{y})_{j_x},~ (\lambda_x)_{j} = (A^T\hat{y})_j - (A^T\hat{y})_{j_x} \geq 0$$

$$ x_{j_x} = 1, ~ x_{j \neq j_x} = 0 $$

$$ \mu_y = (A\hat{x})_{j_y},~ (\lambda_y)_{j} = (A\hat{x})_{j_y} - (A\hat{x})_j \geq 0 $$

$$ y_{j_y} = 1, ~ y_{j \neq j_y} = 0$$

Получается:

$$\text{Err}_{\text{vi}}(z) = \max_{j \in \overline{1, d}} (A\hat{x})_j - \min_{j \in \overline{1, d}} (A^T\hat{y})_j $$

Теоритический шаг для SMP равен:

$$\gamma_k = \dfrac{1}{\sqrt{3} L}$$

Для подбора наилучшего шага рассмотрим:

$$\gamma_k = \dfrac{с}{\sqrt{3} L},$$

затем выберем лучший.

```python
def P(t, z):
    
    summa = 0
    for i in range(d):
        summa += z[i] * np.exp(-t[i])
        
    answer = np.zeros(d)
    for i in range(d):
        answer[i] = 1 / summa * z[i] * np.exp(-t[i])
    
    return answer
```

```python
def Err(x, y):
    return np.max(A @ x) - np.min(A.T @ y)
```

```python
def SMP(grad_f_x, grad_f_y, P, Err, x_0, y_0, gamma, iteration, epsilon=10**-7, c=1):
    
    r_x_now = x_0
    r_y_now = y_0
    summa_w_x = 0
    summa_w_y = 0
    errors = []
    number = [0]
    
    errors.append(Err(x_0, y_0))
    
    for k in trange(iteration, colour='green'):
        
        w_x = P(gamma(k, c) * grad_f_x(r_x_now, r_y_now), r_x_now)
        w_y = P(-gamma(k, c) * grad_f_y(r_x_now, r_y_now), r_y_now)
        r_x_next = P(gamma(k, c) * grad_f_x(w_x, w_y), r_x_now)
        r_y_next = P(-gamma(k, c) * grad_f_y(w_x, w_y), r_y_now)
        
        summa_w_x += w_x
        summa_w_y += w_y
        r_x_now = np.copy(r_x_next)
        r_y_now = np.copy(r_y_next)
        
        errors.append(Err(summa_w_x / (k + 1), summa_w_y / (k + 1)))
        number.append(number[-1] + 4)

        if errors[-1] < epsilon:
            break
    
    return summa_w_x / iteration, summa_w_y / iteration, errors, number
```

```python
def gamma_SMP(k, c=1):
    return c / (np.sqrt(3) * L)
```

__(в)__ Решите билинейную задачу на симплексе с помощью нового метода и методов из Задачи 1. Как будете измерять сходимость? Возможно, в работах, из которых были взяты методы, найдется ответ. Постройте график сходимости. Визуализирйте полученный ответ, как это сделано в файле с лекциями. Сделайте вывод.  

```python
def proj_simplex(x):
    x_sort = sorted(x, reverse=True)
    rho = 0
    summa = x_sort[0]
    summa_ans = x_sort[0]
    for i in range(1, len(x_sort)):
        summa += x_sort[i]
        if x_sort[i] + 1 / (i + 1) * (1 - summa) > 0:
            rho = i
            summa_ans = summa
    lamb = 1 / (rho + 1) * (1 - summa_ans)
    x_answer = np.zeros(len(x_sort))
    for i in range(len(x_answer)):
        x_answer[i] = max(x[i] + lamb, 0)
    return x_answer
```

```python
def DG(grad_f_x, grad_f_y, proj, x_0, y_0, gamma, iteration, epsilon=10**-7, c=1):
    
    x_now = x_0
    y_now = y_0
    errors = []
    number = [0]
    
    errors.append(Err(x_now, y_now))
    
    for k in trange(iteration, colour='green'):
        
        x_next = proj(x_now - gamma(k, c) * grad_f_x(x_now, y_now))
        y_next = proj(y_now + gamma(k, c) * grad_f_y(x_now, y_now))
        
        x_now = x_next
        y_now = y_next
        
        errors.append(Err(x_now, y_now))
        number.append(number[-1] + 2)

        if errors[-1] < epsilon:
            break
            
    return x_now, y_now, errors, number
```

```python
def EG(grad_f_x, grad_f_y, proj, x_0, y_0, gamma, iteration, epsilon=10**-7, c=1):
    
    x_now = x_0
    x_half_now = x_0
    y_now = y_0
    y_half_now = y_0
    errors = []
    number = [0]
    
    errors.append(Err(x_now, y_now))
    
    for k in trange(iteration, colour='green'):
        
        x_half_next = proj(x_now - gamma(k, c) * grad_f_x(x_half_now, y_half_now))
        x_next = proj(x_now - gamma(k, c) * grad_f_x(x_half_now, y_half_now))
        y_half_next = proj(y_now + gamma(k, c) * grad_f_y(x_half_now, y_half_now))
        y_next = proj(y_now + gamma(k, c) * grad_f_y(x_half_now, y_half_now))
        
        x_now = x_next
        x_half_now = x_half_next
        y_now = y_next
        y_half_now = y_half_next
        
        errors.append(Err(x_now, y_now))
        number.append(number[-1] + 4)

        if errors[-1] < epsilon:
            break
            
    return x_now, y_now, errors, number
```

Метод EG (до этого формально был PEG):

$$ x^{k + 1 / 2} = \text{proj} \left( x^k - \gamma_k \nabla_x f(x^{k}, y^{k}) \right)$$

$$ x^{k + 1} = \text{proj} \left( x^k - \gamma_k \nabla_x f(x^{k + 1 / 2}, y^{k + 1 / 2}) \right)$$

$$ y^{k + 1 / 2} = \text{proj} \left( y^k + \gamma_k \nabla_y f(x^{k}, y^{k}) \right)$$

$$ y^{k + 1} = \text{proj} \left( y^k + \gamma_k \nabla_y f(x^{k + 1 / 2}, y^{k + 1 / 2}) \right)$$

Теоритический шаг EG равен:

$$ \gamma_k = \dfrac{1}{4L} $$

Для подбора наилучшего шага рассмотрим:

$$ \gamma_k = \dfrac{c}{4L} $$

затем выберем лучший.

```python
def EG_T(grad_f_x, grad_f_y, proj, x_0, y_0, gamma, iteration, epsilon=10**-7, c=1):
    
    x_now = x_0
    y_now = y_0
    errors = []
    number = [0]
    
    errors.append(np.linalg.norm([x_now - b_x, y_now - b_y], ord=2)**2 / R)
    
    for k in trange(iteration, colour='green'):
        
        x_half = proj(x_now - gamma(k, c) * grad_f_x(x_now, y_now))
        y_half = proj(y_now + gamma(k, c) * grad_f_y(x_now, y_now))
        x_next = proj(x_now - gamma(k, c) * grad_f_x(x_half, y_half))
        y_next = proj(y_now + gamma(k, c) * grad_f_y(x_half, y_half))
        
        x_now = x_next
        y_now = y_next
        
        errors.append(np.linalg.norm([x_now - b_x, y_now - b_y], ord=2)**2 / R)
        number.append(number[-1] + 4)

        if errors[-1] < epsilon:
            break
            
    return x_now, y_now, errors, number
```

Метод RG:

$$ x^{k + 1 / 2} = x^k - (x^{k - 1} - x^{k})$$

$$ x^{k + 1} = \text{proj} \left( x^k - \gamma_k \nabla_x f(x^{k + 1 / 2}, y^{k + 1 / 2}) \right)$$

$$ y^{k + 1 / 2} = y^k - (y^{k - 1} - y^{k})$$

$$ y^{k + 1} = \text{proj} \left( y^k + \gamma_k \nabla_y f(x^{k + 1 / 2}, y^{k + 1 / 2}) \right)$$

Теоритический шаг RG равен:

$$ \gamma_k = \dfrac{1}{4L} $$

Для подбора наилучшего шага рассмотрим:

$$ \gamma_k = \dfrac{c}{4L} $$

затем выберем лучший.

```python
def RG(grad_f_x, grad_f_y, proj, x_0, y_0, gamma, iteration, epsilon=10**-7, c=1):
    
    x_now = x_0
    x_prev = x_0
    y_now = y_0
    y_prev = y_0
    errors = []
    number = [0]
    
    errors.append(np.linalg.norm([x_now - b_x, y_now - b_y], ord=2)**2 / R)
    
    for k in trange(iteration, colour='green'):
        
        x_half = x_now - (x_prev - x_now)
        y_half = y_now - (y_prev - y_now)
        x_next = proj(x_now - gamma(k, c) * grad_f_x(x_half, y_half))
        y_next = proj(y_now + gamma(k, c) * grad_f_y(x_half, y_half))
        
        help_x = np.copy(x_now)
        x_now = np.copy(x_next)
        x_prev = np.copy(help_x)
        help_y = np.copy(y_now)
        y_now = np.copy(y_next)
        y_prev = np.copy(help_y)
        
        errors.append(np.linalg.norm([x_now - b_x, y_now - b_y], ord=2)**2 / R)
        number.append(number[-1] + 2)

        if errors[-1] < epsilon:
            break
            
    return x_now, y_now, errors, number
```

```python
def gamma_RG(k, c=1):
    return c / (4 * L)
```

Метод OG:

$$ x^{k + 1 / 2} = \text{proj} \left( x^k - \gamma_k \nabla_x f(x^{k - 1 / 2}, y^{k - 1 / 2}) \right)$$

$$ x^{k + 1} = x^{k + 1 / 2} - \gamma_k \nabla_x f(x^{k - 1 / 2}, y^{k - 1 / 2}) + \gamma_k \nabla_x f(x^{k + 1 / 2}, y^{k + 1 / 2})$$

$$ y^{k + 1 / 2} = \text{proj} \left( y^k + \gamma_k \nabla_y f(x^{k - 1 / 2}, y^{k - 1 / 2}) \right)$$

$$ y^{k + 1} = y^{k + 1 / 2} + \gamma_k \nabla_y f(x^{k - 1 / 2}, y^{k - 1 / 2}) - \gamma_k \nabla_y f(x^{k + 1 / 2}, y^{k + 1 / 2})$$

Теоритический шаг OG равен:

$$ \gamma_k = \dfrac{1}{4L} $$

Для подбора наилучшего шага рассмотрим:

$$ \gamma_k = \dfrac{c}{4L} $$

затем выберем лучший.

```python
def OG(grad_f_x, grad_f_y, proj, x_0, y_0, gamma, iteration, epsilon=10**-7, c=1):
    
    x_now = x_0
    x_half_now = x_0
    y_now = y_0
    y_half_now = y_0
    errors = []
    number = [0]
    
    errors.append(np.linalg.norm([x_now - b_x, y_now - b_y], ord=2)**2 / R)
    
    for k in trange(iteration, colour='green'):
        
        x_half_next = proj(x_now - gamma(k, c) * grad_f_x(x_half_now, y_half_now))
        y_half_next = proj(y_now + gamma(k, c) * grad_f_y(x_half_now, y_half_now))
        x_next = x_half_next + gamma(k, c) * grad_f_x(x_half_now, y_half_now) - gamma(k, c) * grad_f_x(x_half_next, y_half_next)
        y_next = y_half_next - gamma(k, c) * grad_f_x(x_half_now, y_half_now) + gamma(k, c) * grad_f_x(x_half_next, y_half_next)
        
        x_now = np.copy(x_next)
        x_half_now = np.copy(x_half_next)
        y_now = np.copy(y_next)
        y_half_now = np.copy(y_half_next)
        
        errors.append(np.linalg.norm([x_now - b_x, y_now - b_y], ord=2)**2 / R)
        number.append(number[-1] + 6)

        if errors[-1] < epsilon:
            break
            
    return x_now, y_now, errors, number
```

```python
def gamma_OG(k, c=1):
    return c / (4 * L)
```

Добавьте сходимость на график из пункта (в) Задачи 2, предварительно поменяв ось $x$ (что теперь будете откладывать по оси $x$?).

```python
def EVR(grad_f_x, grad_f_y, proj, Err, x_0, y_0, gamma, alpha, prob, iteration, epsilon=10**-7, c=1):
    
    w_x = x_0
    z_x = x_0
    w_y = y_0
    z_y = y_0
    errors = []
    number = [0]
    
    errors.append(Err(x_0, y_0))
    
    for k in trange(iteration, colour='green'):
        
        z_x_av = alpha * z_x + (1 - alpha) * w_x
        z_y_av = alpha * z_y + (1 - alpha) * w_y
        
        g_x = grad_f_x(w_x, w_y)
        g_y = grad_f_y(w_x, w_y)
        
        z_x_half = proj(z_x_av - gamma(k, prob, c) * g_x)
        z_y_half = proj(z_y_av + gamma(k, prob, c) * g_y)
        
        np.random.seed(k)
        i = np.random.randint(d)
        
        g_x_1 = np.zeros(d)
        g_x_1[i] = grad_f_x(z_x_half, z_y_half)[i]
        g_x_2 = np.zeros(d)
        g_x_2[i] = g_x[i]
        z_x = proj(z_x_av - gamma(k, prob, c) * (g_x + d * g_x_1 - d * g_x_2))
        
        g_y_1 = np.zeros(d)
        g_y_1[i] = grad_f_y(z_x_half, z_y_half)[i]
        g_y_2 = np.zeros(d)
        g_y_2[i] = g_y[i]
        z_y = proj(z_y_av + gamma(k, prob, c) * (g_y + d * g_y_1 - d * g_y_2))
        
        flag = random.choices([0, 1], weights=[prob, 1 - prob])
        
        if flag[0] == 0:
            w_x = np.copy(z_x)
            w_y = np.copy(z_y)
        
        errors.append(Err(w_x, w_y))
        number.append(number[-1] + 2)

        if errors[-1] < epsilon:
            break
    
    return w_x, w_y, errors, number
```

```python
def grad_h(x):
    result = np.log(x) + 1
    return result
```

```python
def MVR(grad_f_x, grad_f_y, Err, x_0, y_0, gamma, alpha, K, iteration, epsilon=10**-14, c=1):
    
    w_x = x_0
    w_y = y_0
    
    w_x_av = grad_h(x_0)
    w_y_av = grad_h(y_0)
    
    z_x = x_0
    z_y = y_0
    
    z_x_half = np.zeros(d)
    z_y_half = np.zeros(d)
    
    errors = []
    number = [0]
    
    errors.append(Err(x_0, y_0))
    
    for s in trange(iteration, colour='green'):
        
        z_x_list = []
        z_y_list = []
        
        for k in range(K):
        
            grad_x = grad_f_x(w_x, w_y)
            grad_y = grad_f_y(w_x, w_y)
        
            z_x_half = alpha * grad_h(z_x) + (1 - alpha) * w_x_av - gamma(k, K, c) * grad_x
            z_y_half = alpha * grad_h(z_y) + (1 - alpha) * w_y_av + gamma(k, K, c) * grad_y
            
            np.random.seed(k * s)
            choice = random.choices(list([t, j] for t in range(d) for t in range(d)), weights=list(np.abs(z_y_half[t] - w_y[t]) / np.linalg.norm(z_y_half - w_y, ord=1) * np.abs(z_x_half[t] - w_x[t]) / np.linalg.norm(z_x_half - w_x, ord=1)  for t in range(d) for t in range(d)))
            
            z_x = np.exp(z_x_half - gamma(k, K, c) * A[choice[0][0]] * np.linalg.norm(z_x_half - w_x, ord=1) * np.sign(z_x_half[choice[0][0]] - w_x[choice[0][0]]) - 1)
            z_y = np.exp(z_y_half + gamma(k, K, c) * A[choice[0][1]] * np.linalg.norm(z_y_half - w_y, ord=1) * np.sign(z_y_half[choice[0][1]] - w_y[choice[0][1]])- 1)
            
            z_x_list.append(z_x)
            z_y_list.append(z_y)
            
        single = np.ones(K)
        w_x = 1 / K * np.array(z_y_list).T @ single
        w_y = 1 / K * np.array(z_x_list).T @ single
        w_x_av = 1 / K * np.array(grad_h(z_x_list)).T @ single
        w_y_av = 1 / K * np.array(grad_h(z_y_list)).T @ single
    
        errors.append(Err(w_x, w_y))
        number.append(number[-1] + 2)

        if errors[-1] < epsilon:
            break
    
    return w_x, w_y, errors, number
```

Теоретические параметры: $\alpha = 1 - \frac{2}{N}$, $\gamma = \frac{0.99 \sqrt{p}}{L}$

```python
def gamma_MVR(k, K, c=1):
    return 0.99 * np.sqrt(1 / K) * c / L
```
