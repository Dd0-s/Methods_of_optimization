## Задача 1

Рассмотрим задачу минимизации:
\begin{equation}
\min_{x \in \mathbb{R}^d} f(x) = \frac{1}{2}x^T A x - b^T x,
\end{equation}
где в качестве матрицы $A$ - случайная матрица с $L =1000$, $\mu = 1$, $d = 100$, $b$ - так же случайный вектор.

__(а)__ Для такой задачи симулируем итерацию SGD:
$$
x^{k+1} = x^k - \gamma (\nabla f(x^k) + \xi^k),
$$
где $\xi^k$ - независимая от $x^k$ случайная величина из нормального распределения $\mathcal{N}(0, 100)$.

Для такого метода с $\gamma = \frac{1}{L}$ постройте график сходимости (критерий $\| \nabla f(x^k)\|^2$). Постройте на этом же графике сходимость честного градиентного спуска c $\gamma = \frac{1}{L}$. Сделайте вывод о характере сходимости нового метода.

В данной задаче:

$$\nabla f(x) = \frac{1}{2} \left(A + A^T \right)x - b$$

В случае симметричной матрицы, которую мы генерируем:

$$\nabla f(x) = Ax - b$$

```python
# Генерация диагональной матрицы
def gen_matrix_diag(d, mu, L, s):
    np.random.seed(s)
    diag = mu + (L - mu) * np.random.sample(d)
    A = np.diag(diag)
    A[0][0] = L
    A[d - 1][d - 1] = mu
    return A


# Генерация матрицы
def gen_matrix(d, mu, L, s):
    np.random.seed(s)
    B = np.random.sample((d, d))
    Q, _ = np.linalg.qr(B)
    C = gen_matrix_diag(d, mu, L, s)
    A = Q @ C @ Q.T
    return A
```

```python
# Целевая функция
def f(x):
    return 1 / 2 * x.T @ A @ x - b.T @ x

# Градиент функции
def grad_f(x):
    return 1 / 2 * (A + A.T) @ x - b
```

```python
def const_gamma(k):
    return 1 / L
```

```python
# Симуляция SGD
def simulation_SGD(grad_f, x_0, gamma, size, iteration, epsilon=10**-7):
    
    x_now = x_0
    errors = []

    errors.append(np.linalg.norm(grad_f(x_now), ord = 2)**2)

    np.random.seed(2)
    
    for k in opt.progress(range(iteration)):
        
        eps = 0
        for i in range(size):
            eps += np.random.normal(0, 10, len(x_0))
        eps = eps / size
        
        x_next = x_now - gamma(k) * (grad_f(x_now) + eps)
        x_now = x_next

        errors.append(np.linalg.norm(grad_f(x_now), ord=2)**2)

        if errors[-1] < epsilon:
            break

    return x_now, errors
```

```python
#Градиентный спуск
def gradient_descent(grad_f, x_0, gamma, iteration, epsilon=10**-7):
    
    x_now = x_0
    errors = []

    errors.append(np.linalg.norm(grad_f(x_now), ord=2)**2)
    
    for k in opt.progress(range(iteration)):
        
        x_next = x_now - gamma(k) * grad_f(x_now)
        x_now = x_next
        
        errors.append(np.linalg.norm(grad_f(x_now), ord=2)**2)
        
        if errors[-1] < epsilon:
            break
            
    return x_now, errors
```

## Задача 2

Рассмотрим задачу логистической регрессии на датасете mushrooms с $\ell_2$-регуляризацией (коэффициент регуляризации равен $\frac{L}{1000}$). Как мы знаем, эта задача имеет вид конечной суммы:
\begin{equation}
\min_{w \in \mathbb{R}^d} f(w) := \frac{1}{n} \sum\limits_{j=1}^n f_j(w) := \frac{1}{n} \sum\limits_{j=1}^n \left[\frac{1}{b} \sum\limits_{i=1}^b f_{ji}(w)\right] := \frac{1}{n} \sum\limits_{j=1}^n \left[\frac{1}{b} \sum\limits_{i=1}^b l (g(w, x_i), y_i)\right],
\end{equation}
где $nb = N$ - полный размер выборки. Решите данную задачу с помощью SGD, SAGA (секция А.5 из [работы](https://arxiv.org/pdf/1905.11261.pdf)), SVRG (секция А.9 из [работы](https://arxiv.org/pdf/1905.11261.pdf)) и SARAH ([работа](https://arxiv.org/pdf/1703.00102.pdf)). 

__(а)__ Оцените $\mu$ и $L$ для данной задачи. Обратите внимание, что теория для SAGA, SVRG, SARAH предполагает, что $L$ - это константа Липшица градиентов $\nabla f_j$. 

Мне кажется, что в правой части пропущен индекс $j$ и должно быть так:

\begin{equation}
    \min_{w \in \mathbb{R}^d} f(w) := \frac{1}{n} \sum\limits_{j=1}^n f_j(w) := \frac{1}{n} \sum\limits_{j=1}^n \left[ \frac{1}{b} \sum\limits_{i=1}^b f_{ji}(w) \right] := \frac{1}{n} \sum\limits_{j=1}^n \left[ \frac{1}{b} \sum\limits_{i=1}^b l (g(w, x_{i + (j - 1)b}, y_{i + (j - 1)b}) \right]
\end{equation}

Воспользуемся результатами из 3 домашнего задания:

$$\nabla^2 f_j(w) = \frac{1}{b} \sum\limits_{i=1}^b \frac{\exp(w^Tx_{i + (j - 1)b}y_{i + (j - 1)b})}{(1 + \exp(w^Tx_{i + (j - 1)b}y_{i + (j - 1)b}))^2} x_{i + (j - 1)b} x_{i + (j - 1)b}^T$$

Где получили:

$$L_j = \frac{1}{4} \max_i \|x_{i + (j - 1)b}\|_2^2, \quad i \in 1 \dots b$$

Тогда:

$$L = \frac{1}{4} \max_i \|x_{i}\|_2^2, \quad i \in 1 \dots N$$

$\mu = 0$, так как:

$$\inf_{z \in \mathbb{R}^d} \left( \frac{\exp(\omega^Tz)}{1 + \exp(\omega^Tz)} \right) = 0$$

```python
#Целевая функция
def f(w):
    s = 0
    for i in range(F.shape[0]):
        s += np.log(1 + np.exp(- w.T @ F[i] * a[i]))
    return 1 / F.shape[0] * s

#Градиент функции 
def grad_f(w):
    s = 0
    for i in range(F.shape[0]):
        s += 1 / (1 + np.exp(w.T @ F[i] * a[i])) * a[i] * F[i]
    return - 1 / F.shape[0] * s

#Частичный градиент
def grad_f_i(w, i):
    B = F[i * b : (i + 1) * b]
    c = a[i * b : (i + 1) * b]
    s = 0
    for i in range(b):
        s += 1 / (1 + np.exp(w.T @ B[i] * c[i])) * c[i] * B[i]
    return - 1 / b * s 

#prox - функция
def prox(w, gamma):
    return max(1 - gamma * L / (1000 * np.linalg.norm(w, ord=2)), 0) * w
```

```python
F = X_train
a = y_train
b = 100
n = a.size // b

seed = 1
np.random.seed(seed)
w_0 = np.random.sample(d)
```

```python
def SGD(grad_f, grad_f_i, x_0, gamma, prox, iteration, m = None, epsilon=10**-7):
    
    x_now = x_0
    errors = []
    times = [0]
    numbers = [0]

    start_time = time.time()
    
    errors.append(np.linalg.norm(grad_f(x_now), ord=2)**2)

    np.random.seed(2)
    
    for k in opt.progress(range(iteration)):
        
        i = np.random.randint(n)
        x_next = prox(x_now - gamma(k) * grad_f_i(x_now, i), gamma(k))
        x_now = x_next

        errors.append(np.linalg.norm(grad_f(x_now), ord=2)**2)
        times.append(time.time() - start_time)
        numbers.append(numbers[-1] + b)
        
        if errors[-1] < epsilon:
            break

    return x_now, errors, times, numbers
```

```python
def SAGA(grad_f, grad_f_i, x_0, gamma, prox, iteration, m = None, epsilon=10**-7):
    
    x_now = x_0
    y_now = n * [x_0]
    errors = []
    times = [0]
    numbers = [0]
    
    start_time = time.time()
    
    summa = 0
    for j in range(n):
        summa += grad_f_i(y_now[j], j)     
    grad = 1 / n * summa
     
    errors.append(np.linalg.norm(grad_f(x_now), ord = 2)**2)
    
    np.random.seed(2)
    
    for k in opt.progress(range(iteration)):
        
        i = np.random.randint(n)
        y_next = y_now[:]
        y_next[i] = x_now
        auxiliary = grad_f_i(y_next[i], i) - grad_f_i(y_now[i], i)
        g = auxiliary + grad
        grad += 1 / n * auxiliary
        y_now = y_next[:]
        x_next = prox(x_now - gamma(k) * g, gamma(k))
        x_now = x_next

        errors.append(np.linalg.norm(grad_f(x_now), ord=2)**2)
        times.append(time.time() - start_time)
        if k == 0:
            numbers.append(numbers[-1] + 2 * b + n * b)
        else:
            numbers.append(numbers[-1] + 2 * b)
        
        if errors[-1] < epsilon:
            break

    return x_now, errors, times, numbers
```

```python
def SVRG(grad_f, grad_f_i, x_0, gamma, prox, iteration, m = 1000, epsilon=10**-7):
    
    x_now = x_0
    y = x_0
    errors = []
    times = [0]
    numbers = [0]
    
    start_time = time.time()
    
    errors.append(np.linalg.norm(grad_f(x_now), ord = 2)**2)

    np.random.seed(2)
    
    for k in opt.progress(range(iteration // m)):
        
        grad = grad_f(y)
        summa = 0
        
        for j in range(m):
            
            i = np.random.randint(n)
            g = grad_f_i(x_now, i) - grad_f_i(y, i) + grad
            x_next = prox(x_now - gamma(k) * g, gamma(k))
            x_now = x_next
            
            summa += x_now

            errors.append(np.linalg.norm(grad_f(x_now), ord=2)**2)
            times.append(time.time() - start_time)
            if j == 0:
                numbers.append(numbers[-1] + 2 * b + n * b)
            else:
                numbers.append(numbers[-1] + 2 * b)
        
        y = 1 / b * summa
        
        if errors[-1] < epsilon:
            break

    return x_now, errors, times, numbers
```

```python
def SARAH(grad_f, grad_f_i, x_0, gamma, prox, iteration, m = 1000, epsilon=10**-7):

    x_now = x_0
    w = x_0
    errors = []
    times = [0]
    numbers = [0]
    
    start_time = time.time()
    
    errors.append(np.linalg.norm(grad_f(x_now), ord = 2)**2)

    np.random.seed(2)
    
    for k in opt.progress(range(iteration // m)):
        
        x_list = []
        
        x_now = w
        x_list.append(x_now)
        
        summa = 0
        for j in range(n):
            summa += grad_f_i(x_now, j)
        y = 1 / n * summa
        
        x_next = prox(x_now - gamma(k)* y, gamma(k))             
        auxiliary = x_now
        x_now = x_next 
        x_prev = auxiliary
        x_list.append(x_now)
        
        for j in range(1, m):
            
            i = np.random.randint(n)
            y += grad_f_i(x_now, i) - grad_f_i(x_prev, i)
            x_next = prox(x_now - gamma(k) * y, gamma(k))
            auxiliary = x_now
            x_now = x_next 
            x_prev = auxiliary
            x_list.append(x_now)

            errors.append(np.linalg.norm(grad_f(x_now), ord=2)**2)

            times.append(time.time() - start_time)
            
            if j == 0:
                numbers.append(numbers[-1] + 2 * b + n * b)
            else:
                numbers.append(numbers[-1] + 2 * b)
        
        i = np.random.randint(m)
        w = x_list[i]
        
        if errors[-1] < epsilon:
            break

    return x_now, errors, times, numbers
```

```python
def accuracy(y_test, X_test, method, grad_f, grad_f_i, x_0, gamma, prox, iteration, step, m = None):
    accuracy = []
    for i in opt.progress(range(1, iteration + 1, step)):
        w, _, _, _ = method(grad_f, grad_f_i, x_0, gamma, prox, i, m)
        y_pred = X_test @ w
        for i in range(len(y_pred)):
            if y_pred[i] > 0:
                y_pred[i] = 1
            else:
                y_pred[i] = -1
        accuracy.append(accuracy_score(y_test, y_pred))
    return accuracy
```

__(г)__ Методы SGD, SAGA, SVRG, SARAH, которые мы рассматривали, предполагают случайный выбор индекса на каждой итерации. Но между тем, на практике обычно делают по-другому и просто последовательно меняют $i_k$ от $1$ до $n$ (а потом заново). В [работе](https://arxiv.org/abs/2006.05988) изучаются такие подходы. Реализуйте Random Reshuffling и Shuffle Once для SGD и SARAH. Сравните новые алгоритмы со старыми версиями SGD и SARAH из пунктов (б) и (в).

```python
def SGD(grad_f, grad_f_i, x_0, gamma, prox, iteration, epsilon=10**-7):
    
    x_now = x_0
    errors = []
    times = [0]
    numbers = [0]

    start_time = time.time()
    
    errors.append(np.linalg.norm(grad_f(x_now), ord=2)**2)

    np.random.seed(3)
    
    for k in opt.progress(range(iteration)):
        
        i = np.random.randint(n)
        x_next = prox(x_now - gamma(k) * grad_f_i(x_now, i), gamma(k))
        x_now = x_next

        errors.append(np.linalg.norm(grad_f(x_now), ord=2)**2)
        times.append(time.time() - start_time)
        numbers.append(numbers[-1] + b)
        
        if errors[-1] < epsilon:
            break

    return x_now, errors, times, numbers
```

```python
def SGD_RR(grad_f, grad_f_i, x_0, gamma, prox, iteration, epsilon=10**-7):
    
    x_now = x_0
    errors = []
    times = [0]
    numbers = [0]

    start_time = time.time()
    
    errors.append(np.linalg.norm(grad_f(x_now), ord=2)**2)
    
    for k in opt.progress(range(iteration // n)):
        
        np.random.seed(k)
        i = list(range(n))
        np.random.shuffle(i)
        
        for j in range(n):
        
            x_next = prox(x_now - gamma(k) * grad_f_i(x_now, i[j]), gamma(k))
            x_now = x_next

            errors.append(np.linalg.norm(grad_f(x_now), ord=2)**2)
            times.append(time.time() - start_time)
            numbers.append(numbers[-1] + b)
        
        if errors[-1] < epsilon:
            break

    return x_now, errors, times, numbers
```

```python
def SARAH(grad_f, grad_f_i, x_0, gamma, prox, iteration, epsilon=10**-7):

    x_now = x_0
    w = x_0
    errors = []
    times = [0]
    numbers = [0]
    
    start_time = time.time()
    
    errors.append(np.linalg.norm(grad_f(x_now), ord = 2)**2)
    
    for k in opt.progress(range(iteration // n)):
        
        x_list = []
        
        x_now = w
        x_list.append(x_now)
        
        summa = 0
        for j in range(n):
            summa += grad_f_i(x_now, j)
        y = 1 / n * summa
        
        x_next = prox(x_now - gamma(k)* y, gamma(k))             
        auxiliary = x_now
        x_now = x_next 
        x_prev = auxiliary
        x_list.append(x_now)
        numbers.append(numbers[-1] + n * b)
        
        for j in range(n):
            
            np.random.seed(k)
            i = np.random.randint(n)
            y += grad_f_i(x_now, i) - grad_f_i(x_prev, i)
            x_next = prox(x_now - gamma(k) * y, gamma(k))
            auxiliary = x_now
            x_now = x_next 
            x_prev = auxiliary
            x_list.append(x_now)

            errors.append(np.linalg.norm(grad_f(x_now), ord=2)**2)

            times.append(time.time() - start_time)
            
            if j == 0:
                numbers[-1] += 2 * b
            else:
                numbers.append(numbers[-1] + 2 * b)
        
        index = np.random.randint(n)
        w = x_list[index]
        
        if errors[-1] < epsilon:
            break

    return x_now, errors, times, numbers
```

__(д)__ Методы SVRG и SARAH имеют loopless версии: L-SVRG (секция А.10 из [работы](https://arxiv.org/pdf/1905.11261.pdf) и PAGE ([работа](https://proceedings.mlr.press/v139/li21a.html)), соотвественно. Реализуйте их. Как будете выбирать вероятности в этих алгоритмах? Сравните новые алгоритмы со старыми версиями SVRG и SARAH из пунктов (б) и (в).

```python
def SVRG(grad_f, grad_f_i, x_0, gamma, prox, iteration, prob = None, epsilon=10**-7):
    
    x_now = x_0
    y = x_0
    errors = []
    times = [0]
    numbers = [0]
    
    start_time = time.time()
    
    errors.append(np.linalg.norm(grad_f(x_now), ord = 2)**2)

    np.random.seed(2)
    
    for k in opt.progress(range(iteration // n)):
        
        grad = grad_f(y)
        summa = 0
        
        for j in range(n):
            
            i = np.random.randint(n)
            g = grad_f_i(x_now, i) - grad_f_i(y, i) + grad
            x_next = prox(x_now - gamma(k) * g, gamma(k))
            x_now = x_next
            
            summa += x_now

            errors.append(np.linalg.norm(grad_f(x_now), ord=2)**2)
            times.append(time.time() - start_time)
            if j == 0:
                numbers.append(numbers[-1] + 2 * b + n * b)
            else:
                numbers.append(numbers[-1] + 2 * b)
        
        y = 1 / b * summa
        
        if errors[-1] < epsilon:
            break

    return x_now, errors, times, numbers
```

```python
def L_SVRG(grad_f, grad_f_i, x_0, gamma, prox, iteration, prob = 1 / n, epsilon=10**-7):
    
    x_now = x_0
    w = x_0
    g = np.zeros(n)
    flag = 0
    
    errors = []
    times = [0]
    numbers = [0]

    start_time = time.time()
    
    errors.append(np.linalg.norm(grad_f(x_now), ord=2)**2)
    
    for k in opt.progress(range(iteration)):
        
        np.random.seed(k)
        i = np.random.randint(n)
        
        if flag == 0:
            grad = grad_f(w)
        
        g = grad_f_i(x_now, i) - grad_f_i(w, i) + grad[i]
        x_next = prox(x_now - gamma(k) * grad_f_i(x_now, i), gamma(k))
        flag = random.choices([0, 1], weights=[prob, 1 - prob])
        if flag == 0:
            w = x_next
            
        x_now = x_next

        errors.append(np.linalg.norm(grad_f(x_now), ord=2)**2)
        times.append(time.time() - start_time)
        if flag == 0:
            numbers.append(numbers[-1] + 2 * b + n * b)
        else:
            numbers.append(numbers[-1] + 2 * b)
        
        if errors[-1] < epsilon:
            break

    return x_now, errors, times, numbers
```

```python
def SARAH(grad_f, grad_f_i, x_0, gamma, prox, iteration, prob = None, epsilon=10**-7):

    x_now = x_0
    w = x_0
    errors = []
    times = [0]
    numbers = [0]
    
    start_time = time.time()
    
    errors.append(np.linalg.norm(grad_f(x_now), ord = 2)**2)
    
    for k in opt.progress(range(iteration // n)):
        
        x_list = []
        
        x_now = w
        x_list.append(x_now)
        
        summa = 0
        for j in range(n):
            summa += grad_f_i(x_now, j)
        y = 1 / n * summa
        
        x_next = prox(x_now - gamma(k)* y, gamma(k))             
        auxiliary = x_now
        x_now = x_next 
        x_prev = auxiliary
        x_list.append(x_now)
        numbers.append(numbers[-1] + n * b)
        
        for j in range(n):
            
            np.random.seed(k)
            i = np.random.randint(n)
            y += grad_f_i(x_now, i) - grad_f_i(x_prev, i)
            x_next = prox(x_now - gamma(k) * y, gamma(k))
            auxiliary = x_now
            x_now = x_next 
            x_prev = auxiliary
            x_list.append(x_now)

            errors.append(np.linalg.norm(grad_f(x_now), ord=2)**2)

            times.append(time.time() - start_time)
            
            if j == 0:
                numbers[-1] += 2 * b
            else:
                numbers.append(numbers[-1] + 2 * b)
        
        index = np.random.randint(n)
        w = x_list[index]
        
        if errors[-1] < epsilon:
            break

    return x_now, errors, times, numbers
```

```python
def PAGE(grad_f, grad_f_i, x_0, gamma, prox, iteration, prob = 1 / (1 + b), epsilon=10**-7):

    x_now = x_0
    errors = []
    times = [0]
    numbers = [0]
    
    np.random.seed(3)
    i = np.random.randint(n)
    g = grad_f_i(x_now, i)
    
    start_time = time.time()
    errors.append(np.linalg.norm(grad_f(x_now), ord = 2)**2)
    numbers.append(numbers[-1] + n * b)
    
    for k in opt.progress(range(iteration)):
        
        x_list = []
        x_list.append(x_now)
         
        x_next = prox(x_now - gamma(k)* g, gamma(k))             
         
        flag = random.choices([0, 1], weights=[prob, 1 - prob])
        np.random.seed(k)
        i = np.random.randint(n)
        if flag[0] == 0:
            g = grad_f_i(x_next, i)
        else:
            g += grad_f_i(x_next, i) - grad_f_i(x_now, i)
            
        x_now = x_next

        errors.append(np.linalg.norm(grad_f(x_now), ord=2)**2)
        times.append(time.time() - start_time)
        
        if k == 0:
            if flag == 0:
                numbers[-1] += b
            else:
                numbers[-1] += 2 * b
        else:
            if flag == 0:
                numbers.append(numbers[-1] + b)
            else:
                numbers.append(numbers[-1] + 2 * b)
                   
        if errors[-1] < epsilon:
            break

    return x_now, errors, times, numbers
```

```python
def accuracy(y_test, X_test, method, grad_f, grad_f_i, x_0, gamma, prox, iteration, prob, step):
    accuracy = []
    for i in opt.progress(range(1, iteration + 1, step)):
        w, _, _, _ = method(grad_f, grad_f_i, x_0, gamma, prox, i, prob)
        y_pred = X_test @ w
        for i in range(len(y_pred)):
            if y_pred[i] > 0:
                y_pred[i] = 1
            else:
                y_pred[i] = -1
        accuracy.append(accuracy_score(y_test, y_pred))
    return accuracy
```

```python
accuracy_1 = accuracy(y_test, X_test, SVRG, grad_f, grad_f_i, w_0, gamma_SGD, prox, 300, None, 100)
accuracy_2 = accuracy(y_test, X_test, L_SVRG, grad_f, grad_f_i, w_0, gamma_SGD, prox, 300, 1 / n, 100)
accuracy_3 = accuracy(y_test, X_test, SARAH, grad_f, grad_f_i, w_0, gamma_SGD, prox, 300, None, 100)
accuracy_4 = accuracy(y_test, X_test, PAGE, grad_f, grad_f_i, w_0, gamma_SARAH, prox, 300, 1 / (1 + b), 100)
```
